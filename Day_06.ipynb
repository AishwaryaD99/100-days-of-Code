```python
from bs4 import BeautifulSoup
import requests
```

```python
req = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Analyst&txtLocation=").text
```

```python
soup = BeautifulSoup(req, "lxml")
```

```python
print(soup.prettify())
```

```python
print(soup.get_text())
```

```python
job = soup.find('li',class_ = "clearfix job-bx wht-shd-bx")
```

```python
comp_name = job.find('h3',class_ = "joblist-comp-name").text.replace(" ","")
```

```python
print(comp_name)
```

```python
skills = job.find("span", class_= "srp-skills").text
```

```python
print(skills)
```

```python
publ_date = job.find("span",class_ = 'sim-posted').span.text
```

```python
print(publ_date)
```

```python
print(f'''
Company Name : {comp_name}
Required Skills : {skills}
''')
```

```python
jobs = soup.find_all('li',class_ = "clearfix job-bx wht-shd-bx")
```

```python
for j in jobs :
    publ_date = j.find("span",class_ = 'sim-posted').span.text.replace(" ","")
    if "few" in publ_date :
        comp_name = j.find('h3',class_ = "joblist-comp-name").text.replace(" ","")
        skills = j.find("span", class_= "srp-skills").text.replace(" ","")
        print(f'''
        Company Name : {comp_name}
        Required Skills : {skills}
        ''')
        print(" ")
```
